# 1.1.2 Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
cols = ['Customer','state','CustomerLifetimeValue','Response','Coverage','Education','EffectiveToDate','EmploymentStatus','Gender','Income','LocationCode','MaritalStatus','MonthlyPremiumAuto','MonthsSinceLastClaim','MonthsSincePolicyInception','NumberofOpenComplaints','NumberofPolicies','PolicyType','Policy','RenewOfferType','SalesChannel','ClaimAmount','VehicleClass','VehicleSize']
print(cols)
```

# 1.1.3 One hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
obj = OneHotEncoder(sparse_output = False, dtype = int)
new_col = obj.fit_transform(df[['Gender']])
print(type(new_col))
```

```python
encoder_df = pd.DataFrame(new_col, columns = obj.get_feature_names_out(['Gender']))
encoder_df 
```

```python
df = pd.concat([df,encoder_df],axis = 1)
df
```

```python
df = df.drop('Gender', axis = 1)
df
```

###    1.2 Feature Scaling:

# 1.2.1 Min-Max scaling

```python
from sklearn.preprocessing import MinMaxScaler
obj = MinMaxScaler(feature_range=(0,1))
df['CustomerLifetimeValue'] = obj.fit_transform(df[['CustomerLifetimeValue']])
df
```

# 1.2.2 Standardization

```python
from sklearn.preprocessing import StandardScaler
obj = StandardScaler()
df['ClaimAmount'] = obj.fit_transform(df[['ClaimAmount']])
df
```

# 1.2.3 Robust scaling

```python
from sklearn.preprocessing import RobustScaler
obj = RobustScaler()
df['Income'] = obj.fit_transform(df[['Income']])
df
```

# 2.2 Chi-square method

```python
from scipy.stats import chi2_contingency
# create a categorical matrix for chi-square method
contigency_matrix = pd.crosstab(df['Income'],df['EmploymentStatus'])
contigency_matrix
```

```python
_ , p_value, _ , _ = chi2_contingency(contigency_matrix)
p_value
```

# 3. Feature Balancing
###    3.1 Oversampling:
####        3.1.1 SMOTE
####        3.1.2 Random oversampling


# 3.1 Oversampling

```python
df.EmploymentStatus.value_counts().plot(kind = 'pie', autopct = '%0.0f%%')
```

```python
# RandomOverSample
from imblearn.over_sampling import RandomOverSampler, SMOTE
x = df.drop('EmploymentStatus', axis = 1)
y = df.loc[:, ['EmploymentStatus']]

obj_ros = RandomOverSampler(sampling_strategy = 'auto')
x_ros, y_ros = obj_ros.fit_resample(x,y)

y_ros.value_counts()
```

```python
y.value_counts()
```

```python
# SMOTE
obj_smote = SMOTE(sampling_strategy = 'auto')
x_smote, y_smote = obj_smote.fit_resample(x,y)
```

```python
y_smote.value_counts()
```

## 3.2 Undersampling:
####        3.2.1 Random under sampler
####        3.2.2 Edited Nearest Neighbour


```python
# RandomUnderSampler
from imblearn.under_sampling import RandomUnderSampler
obj_rus = RandomUnderSampler(sampling_strategy = 'not majority')
x_rus, y_rus = obj_rus.fit_resample(x,y)

y_rus.value_counts()
```

```python
# EditedNearestNeighbours
from imblearn.under_sampling import EditedNearestNeighbours
obj_enn = EditedNearestNeighbours(n_neighbors = 5)
x_enn, y_enn = obj_enn.fit_resample(x,y)

y_enn.value_counts()
```
